```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
from statsmodels.graphics.tsaplots import plot_acf
```

```python
 
# Load the dataset
data = pd.read_excel('C:\\Users\\tajes\\Downloads\\BakeryData2024_Vilnius.xlsx')

```

```python
  #Task 5: Descriptive Analysis and Visualization

# (a) Calculate descriptive statistics
descriptive_stats = data.describe()
print(descriptive_stats)
```

```python

# (b) Box plots for daily demand for each store
data.boxplot(column=['main street A', 'main street B', 'station A', 'station B'])
plt.title('Box Plot of Daily Demand for Each Store')
plt.ylabel('Daily Demand (0.25 kg)')
plt.show()
```

```python

# (c) Histograms for daily demand for each store
data.hist(column=['main street A', 'main street B', 'station A', 'station B'], bins=20, figsize=(10, 6))
plt.suptitle('Histograms of Daily Demand for Each Store')
plt.show()
```

```python

# (d) Time series plots for daily demand
dates = pd.date_range(start='2016-05-11', end='2024-05-29')
plt.figure(figsize=(12, 8))
for column in data.columns:
    plt.plot(dates, data[column], label=column)
plt.title('Time Series Plot of Daily Demand')
plt.xlabel('Date')
plt.ylabel('Daily Demand (0.25 kg)')
plt.legend()
plt.show()
```
